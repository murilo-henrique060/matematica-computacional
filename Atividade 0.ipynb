```python
import colorsys
import matplotlib.pyplot as plt
import numpy as np

colors = (
    'blue',
    'green',
    'red',
    'cyan'
)

def plot(*vectors, title=None, size=10, figsize=6):
  plt.figure(figsize=(figsize, figsize))

  if (title is not None):
    plt.title(title)

  plt.grid(True)

  plt.axhline(0, color='black',linewidth=0.5)
  plt.axvline(0, color='black',linewidth=0.5)

  half = size/2

  plt.axis([-half,half,-half,half])

  for i, vector in enumerate(vectors):
    color = i % len(colors)
    plt.arrow(0,0, vector[0], vector[1], head_width=0.2, head_length=0.3, fc=colors[color], ec=colors[color])

  plt.show()
```

Para obter as matrizes de transformação de rotação e shear começamos com os vetores unitários i e j, com valores (1,0) e (0,1), respectivamente.

```python
i = np.array([1, 0])
j = np.array([0, 1])

plot(i, j, title="Vetores Base Unitários")
```

Para a matriz de rotação, aplicamos a transformação de rotação em 90° no sentido horário, obtendo novos i e j com valores (0,-1) e (1,0), e a partir destes vetores podemos criar uma matriz que representa a transformação.

$$
M_R =
\begin{bmatrix}
   0 & 1 \\
   -1 & 0
\end{bmatrix}
$$

```python
rotation_matrix = np.array([
    [0, 1],
    [-1, 0]
])

rotation_i = np.dot(rotation_matrix, i)
rotation_j = np.dot(rotation_matrix, j)

plot(rotation_i, rotation_j, title="Vetores Unitários Rotacionados")
```

Para a matriz de shear no eixo x, mantemos o vetor i em sua posição original e movemos o vetor j horizontalmente, assim obtendo (-1,1), e a partir desses vetores podemos criar a matriz de transformação.

$$
M_S =
\begin{bmatrix}
   1 & -1 \\
   0 & 1
\end{bmatrix}
$$

```python
shear_matrix = np.array([
    [1, -1],
    [0, 1]
])

shear_i = np.dot(shear_matrix, i)
shear_j = np.dot(shear_matrix, j)

plot(shear_i, shear_j, title="Vetores Unitários Cisalhados")
```

Com as duas matrizes de transformação, podemos começar com dois vetores:

```python
vector_1 = np.array([1, 2])
vector_2 = np.array([2, -1])

plot(vector_1, vector_2, title=f'Vetores arbitrários ({vector_1[0]:.0f}, {vector_1[1]:.0f}) e ({vector_2[0]:.0f}, {vector_2[1]:.0f})')
```

Depois aplicamos a rotação multiplicando a matriz de rotação pelo vetor

```python
rotated_vector_1 = np.dot(rotation_matrix, vector_1)
rotated_vector_2 = np.dot(rotation_matrix, vector_2)

plot(rotated_vector_1, rotated_vector_2, title=f'Rotated Vector ({rotated_vector_1[0]:.0f}, {rotated_vector_1[1]:.0f}) e ({rotated_vector_2[0]:.0f}, {rotated_vector_2[1]:.0f})')
```

Após isso, multiplicamos o resultado pela matriz de shear

```python
shear_rotated_vector_1 = np.dot(shear_matrix, rotated_vector_1)
shear_rotated_vector_2 = np.dot(shear_matrix, rotated_vector_2)

plot(shear_rotated_vector_1, shear_rotated_vector_2, title=f'Shear Rotated Vector ({shear_rotated_vector_1[0]:.0f}, {shear_rotated_vector_1[1]:.0f}) e ({shear_rotated_vector_2[0]:.0f}, {shear_rotated_vector_2[1]:.0f})')
```
